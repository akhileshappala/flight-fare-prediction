```python
import numpy as np
import pandas as pd

from sklearn import preprocessing, svm, model_selection
from sklearn.linear_model import LinearRegression, ElasticNet
from sklearn.ensemble import RandomForestRegressor

import statsmodels.api as sm
from statsmodels.formula.api import ols                                        
```

```python
from IPython.display import HTML, display

data = pd.read_csv("data.csv")[['ITIN_ID', 'ORIGIN', 'ORIGIN_AIRPORT_ID', 'ORIGIN_STATE_ABR', 'ROUNDTRIP', 'ITIN_YIELD', 'ITIN_FARE', 'DISTANCE']]

# data_model = ols("""ITIN_FARE ~ ITIN_YIELD + ORIGIN + ORIGIN_STATE_ABR + ROUNDTRIP + DISTANCE""", data=data).fit()


data_model = ols("""ITIN_FARE ~ ITIN_YIELD 
                                            + ORIGIN 
                                            + ORIGIN_STATE_ABR
                                            + ROUNDTRIP
                                            + DISTANCE""", data=data).fit()


model_summary = data_model.summary()

HTML(
model_summary\
.as_html()\
.replace(' Adj. R-squared: ', ' Adj. R-squared: ')\
.replace('coef', 'coef')\
.replace('std err', 'std err')\
.replace('P>|t|', 'P>|t|')\
.replace('[95.0% Conf. Int.]', '[95.0% Conf. Int.]')
)
```
